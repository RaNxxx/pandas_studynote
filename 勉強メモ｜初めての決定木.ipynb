```python
import pandas as pd
import numpy as np
from sklearn import tree
import graphviz
import math
```

```python
hr_data_train = pd.read_csv("ibm_data_train.csv")
hr_data_test = pd.read_csv("ibm_data_test.csv")

hr_data_train
```

```python
MAX_DEPTH = 6
hr_regressor = tree.DecisionTreeRegressor(max_depth=MAX_DEPTH)
```

```python
#X_trainã‚’ä½œã‚‹
X_train = pd.DataFrame()

#æ•°å­—ã®ã‚«ãƒ©ãƒ ã®ã„ã¿æ®‹ã™
for i,item in hr_data_train.iteritems():
    if type(item[0]) == np.int64:
        #X_train = pd.concat([X_train, item], axis=1)
        X_train[i] = item

#TotalWorkingYearsã‚’é™¤ã
X_train = X_train.drop(["TotalWorkingYears"], axis=1)

X_train
```

```python
#y_trainã‚’ä½œæˆ
y_train = hr_data_train["TotalWorkingYears"]
```

```python
#å­¦ç¿’ã™ã‚‹
#yã¯æ•™å¸«ãƒ‡ãƒ¼ã‚¿ã€
#è¡Œåˆ—ã¯å¤§æ–‡å­—ã§æ›¸ãã€ãƒ™ã‚¯ãƒˆãƒ«ã¯å°æ–‡å­—ã§æ›¸ã
#äºˆæ¸¬ã—ãŸã„ã‚‚ã®ã¯target(yã€ä¸€æ¬¡å…ƒã®ãƒ™ã‚¯ãƒˆãƒ«ã€å°æ–‡å­—ã§æ›¸ã)ã¨å‘¼ã¶ã€äºˆæ¸¬ã«ä½¿ã†ãƒ‡ãƒ¼ã‚¿ã¯feature(Xã€è¡Œåˆ—ã§å¤§æ–‡å­—ã§æ›¸ã)ã¨å‘¼ã¶
hr_regressor.fit(X_train, y_train)
```

```python
#testã®ã‚»ãƒƒãƒˆã‚’ä½œæˆã™ã‚‹
#X_testã‚’ä½œã‚‹
X_test = pd.DataFrame()

#æ•°å­—ã®ã‚«ãƒ©ãƒ ã®ã„ã¿æ®‹ã™
for i,item in hr_data_test.iteritems():
    if type(item[0]) == np.int64:
        X_test[i] = item

#TotalWorkingYearsã‚’é™¤ã
X_test = X_test.drop(["TotalWorkingYears"], axis=1)

X_test
```

```python
#äºˆæ¸¬ã™ã‚‹
hr_predict = hr_regressor.predict(X_test)
hr_predict
```

```python
df_predict = pd.DataFrame(hr_predict)
predict_result = df_predict.rename(columns={0:"Predict_Result"})
predict_result
```

```python
y_totalworkingyears = hr_data_test["TotalWorkingYears"]
y_totalworkingyears = pd.DataFrame(y_totalworkingyears)
```

```python
#predictã®çµæžœã¨ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã®å¹³å‡äºŒä¹—èª¤å·®ã‚’æ±‚ã‚ã‚‹
#ã²ã‚ãå…ˆç”Ÿã®ã”æŒ‡ç¤ºã«ã‚ˆã‚Šã€ã“ã“ã§ã¯ã€ã‚ãˆã¦å¹³å‡äºŒä¹—èª¤å·®ã‚’æ±‚ã‚ã‚‹é–¢æ•°ã‚’è‡ªå‰ã§ä½œæˆã™ã‚‹
compassion_df = pd.concat([y_totalworkingyears, predict_result], axis=1)
compassion_df
```
