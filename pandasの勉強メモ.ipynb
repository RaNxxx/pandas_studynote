```python
#Numpyのarrayを使う
array1 = np.array([[1,2,3],[23,21,3],[3,2,2]])
array1
```

```python
#Numpyのarrayをデータフレームに変換する
pd.DataFrame(array1)
```

```python
#columnsを定義することもできる
df1 = pd.DataFrame(array1, columns = ["A","B","C"])
```

```python
array2 = np.array([[10,20,30],[230,210,30],[30,20,20]])
array2
```

```python
df2 = pd.DataFrame(array2, columns = ["A", "B", "C"])
```

```python
#データフレームを縦に結合することができる
pd.concat([df1, df2])
```

```python
#0,1,2が重複し、それを連番したい時は、ignore_indexが使える
pd.concat([df1, df2], ignore_index = True)
```

```python
#行を繋げたい時はconcatを使う
df3 = pd.concat([df1, df2], axis = 1)
df3
```

```python
df3.columns = ["A","B","C","D","E","F"]
df3
```

```python
#カラム名を変更
df3.rename(columns={"A":"a", "D":"xxx"})
```

```python
#indexを変更
df3.rename(index={0:1001, 2:1011})
```

```python
#ただし、renameの時はrenameされたデータフレームを返しているだけ
#したがって、元のデータフレームを変更された訳ではない
df3
```

```python
#データフレームの結合
customer_df = pd.DataFrame([["001","Hiroki"],["002","Rarurun"]],columns=["customer_id", "name"])
customer_df
```

```python
order_df = pd.DataFrame([["001","Game"],
                         ["001","Food"],
                         ["002","Plant"]],
                         columns=["customer_id", "purchase_item"])
order_df
```

```python
pd.merge(customer_df, order_df, how="inner", on="customer_id")
```

```python
#キーが異なる場合のデータフレーム結合
employee_df = pd.DataFrame([["Hiroki","Miyagi"],["Rarurun","Sichuan"],["Puuu","Ikebukuro"]],columns=["name","hometown"])
employee_df
```

```python
office_df = pd.DataFrame([["Tokyo","Tokyo-1-2-3"],["Sichuan","Tokyo-2-1-2"],["Ikebukuro","Tokyo-2-2-2"]],columns=["city","address"])
office_df
```

```python
pd.merge(employee_df, office_df, how="inner", left_on="hometown",right_on="city")
```

```python
#外部結合
product_df = pd.DataFrame([["P-001","Orange"],
                          ["P-002","Pink"],
                          ["P-003","Black"]],
                          columns=["product","color"])
product_df
```

```python
#mergeの異なる処理を認識するために、product_dfにない項目P-005を作成する
stock_df = pd.DataFrame([["P-001",10],
                         ["P-002",20],
                         ["P-005",33]],
                         columns=["product","number"])
stock_df
```

```python
#innerは両方のデータフレームにあるものだけを結合する
pd.merge(product_df, stock_df, how="inner", on="product")
```

```python
#leftは1個目のデータフレームにあるものに基づき、2個目で重複したものを結合する
pd.merge(product_df, stock_df, how="left", on="product")
```

```python
#rightはその逆である
pd.merge(product_df, stock_df, how="right", on="product")
```

```python
#outerはいずれのデータフレームにあるものを全部結合する
pd.merge(product_df, stock_df, how="outer", on="product")
```

```python
#csvファイルとして出力
df_output = pd.merge(product_df, stock_df, how="outer", on="product")
df_output.to_csv("merge_product.csv")
```
